```python
import pandas as pd
import numpy as np

import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})

one_hot = np.zeros((len(lst), 2))
for i in range(len(lst)):
    if lst[i] == 'robot':
        one_hot[i][0] = 1
    else:
        one_hot[i][1] = 1

one_hot_df = pd.DataFrame(one_hot, columns=['robot', 'human'])
one_hot_df.head()
```
